# Data Cleaning and Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Boston data set, csv was downloaded and saved as boston_housing.csv
boston = pd.read_csv("boston_housing.csv")

# Check the first 10 records 
boston.head(10)
```

```python
# Total number of the records
boston.shape
```

```python
# Removing the columns CHAS, NOX, B, LSTAT and create new dataframe
boston_new = boston.drop(['CHAS', 'NOX', 'B', 'LSTAT'], axis=1)
```

```python
# The last seven records of new dataframe
boston_new.tail(7)
```

```python
# Plot histogram with all the columns in the new dataframes one at a time
figure, axis = plt.subplots(5,2, tight_layout = True)
figure.set_figwidth(10)
figure.set_figheight(15)
axis[0,0].hist(boston_new["CRIM"])
axis[0,0].set_title('Plot of CRIM')
axis[0,1].hist(boston_new["ZN"])
axis[0,1].set_title('Plot of ZH')
axis[1,0].hist(boston_new["INDUS"])
axis[1,0].set_title('Plot of INDUS')
axis[1,1].hist(boston_new["RM"])
axis[1,1].set_title('Plot of RM')
axis[2,0].hist(boston_new["AGE"])
axis[2,0].set_title('Plot of AGE')
axis[2,1].hist(boston_new["DIS"])
axis[2,1].set_title('Plot of DIS')
axis[3,0].hist(boston_new["RAD"])
axis[3,0].set_title('Plot of RAD')
axis[3,1].hist(boston_new["TAX"])
axis[3,1].set_title('Plot of TAX')
axis[4,0].hist(boston_new["PTRATIO"])
axis[4,0].set_title('Plot of PTRATIO')
axis[4,1].hist(boston_new["PRICE"])
axis[4,1].set_title('Plot of PRICE')
```

```python
# Plot histograms in the new data frame using a for loop

for c in boston_new.columns:
    plt.title(f"Plot of {c}", fontsize=15)
    plt.hist(boston_new[c],bins=20)
    plt.show()
```

```python
# Create a scatter plot of the crime rate vs price. Also title the histogram 

plt.scatter(boston_new['CRIM'], boston_new['PRICE'])
plt.title("Crime vs Price")
plt.xlabel("Crime")
plt.ylabel("Price")
plt.show()
```

```python
# Create a plot with the log10 of the crime vs. price. Also title and label the graph
plt.scatter(np.log10(boston_new["CRIM"]), boston_new["PRICE"])
plt.title("Scatter Plot of Log 10 Crime vs Price")
plt.xlabel("Crime")
plt.ylabel("Price") 
plt.show()
```
